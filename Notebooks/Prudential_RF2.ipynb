```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('../Data/train.csv')
```

```python
data.info()
```

```python
pd.set_option('display.max_columns', None)
data.head(3)
```

```python
train_dummy_data=pd.get_dummies(data)
```

```python
train_data=train_dummy_data.copy()
```

## Null Values

```python
pd.set_option('display.max_rows', None)
train_data.isnull().sum()
```

```python
train_data.drop('Id', axis=1, inplace=True)
train_data.head()
```

## new dataframe to fill NaN's if missing less than 600 values.  Drop if more

```python
D_train=train_data.copy()
```

```python
D_train['Employment_Info_1'].fillna(0, inplace=True)
```

```python
D_train.drop(['Employment_Info_4', 'Employment_Info_6', 'Insurance_History_5', 'Family_Hist_2',
             'Family_Hist_3', 'Family_Hist_4', 'Family_Hist_5', 'Medical_History_1', 'Medical_History_10',
             'Medical_History_15', 'Medical_History_24', 'Medical_History_32'], axis=1, inplace=True)
```

## Random Forest Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
train_X=D_train.drop('Response', axis=1)
train_y=D_train['Response']
```

```python
D_Tree=DecisionTreeClassifier()

D_Tree.fit(train_X, train_y)
D_Tree.score(train_X, train_y)
```

## Bring in Test Data

```python
test_data=pd.read_csv('../Data/test.csv')
```

```python
test_data.head(2)
```

```python
## Put aside ID's for submission
test_IDs=test_data['Id']
```

#### get dummies

```python
test_dummy_data=pd.get_dummies(test_data)
```

### Drop columns that were dropped in training data

```python
test_dummy_data.drop(['Id','Employment_Info_4', 'Employment_Info_6', 'Insurance_History_5', 'Family_Hist_2',
             'Family_Hist_3', 'Family_Hist_4', 'Family_Hist_5', 'Medical_History_1', 'Medical_History_10',
             'Medical_History_15', 'Medical_History_24', 'Medical_History_32'], axis=1, inplace=True)
```

#### Fill NaN's with 0

```python
test_dummy_data.isnull().sum()
```

```python
test_dummy_data.fillna(0, inplace=True)
```

### Make Predictions

```python
test_X=test_dummy_data
```

```python
predictions=D_Tree.predict(test_X)
```

```python
Test_Submission=pd.DataFrame({
    'Id': test_IDs,
    'Response': predictions
})
```

```python
Test_Submission.Response.hist()
```

```python
Test_Submission.to_csv(r'../Submissions/D_Tree_Submission1.csv', index=False, header=True)
```
