```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('../Data/train.csv')
```

```python
data.info()
```

```python
pd.set_option('display.max_columns', None)
data.head(3)
```

```python
train_dummy_data=pd.get_dummies(data)
```

```python
train_data=train_dummy_data.copy()
```

```python
train_data.drop("Id", axis=1, inplace=True)
train_data.head(2)
```

## Null Values

```python
pd.set_option('display.max_rows', None)
train_data.isnull().sum()
```

## new dataframe to fill NaN's with 0's

```python
Z_train=train_data.copy()
```

```python
# def fill_median(column):
#     median=column.median()
#     column.fillna(median, inplace=True)
    
```

```python
Z_train.fillna(0, inplace=True)
```

```python
Z_train.isnull().sum()[:3]
```

## Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
train_X=Z_train.drop('Response', axis=1)
train_y=Z_train['Response']
```

```python
Z_GB=GradientBoostingClassifier(n_estimators=500)

Z_GB.fit(train_X, train_y)
Z_GB.score(train_X, train_y)
```

## Bring in Test Data

```python
test_data=pd.read_csv('../Data/test.csv')
```

```python
test_data.head(2)
```

```python
## Put aside ID's for submission
test_IDs=test_data['Id']
```

#### get dummies

```python
test_dummy_data=pd.get_dummies(test_data)
```

#### Fill NaN's with 0

```python
test_dummy_data.fillna(0, inplace=True)
```

### Make Predictions

```python
test_X=test_dummy_data.drop('Id', axis=1)
```

```python
predictions=Z_GB.predict(test_X)
```

```python
Test_Submission=pd.DataFrame({
    'Id': test_IDs,
    'Response': predictions
})
```

```python
Test_Submission.Response.hist()
```

```python
Test_Submission.to_csv(r'../Submissions/Z_GB_Submission2.csv', index=False, header=True)
```
